```python
# Exploratory Computational Analysis (Pilot)
# Purpose: test computational feasibility of a sentence-level cognitive friction framework

# Synthetic SOP sentences (S1–S10)

sentences = [
    "If the test result exceeds the acceptance criteria, a retest shall be performed in accordance with SOP-QC-012.",
    "The deviation shall be reported within 24 hours of identification using the designated deviation reporting form.",
    "However, when additional testing is required, the investigation timeline may be extended as appropriate.",
    "This step is not applicable to stability samples collected for long-term studies.",
    "Where applicable, the analyst should review historical data prior to initiating the investigation.",
    "The investigation shall include an assessment of potential laboratory errors, equipment performance, and sample handling history.",
    "Unless otherwise specified, all raw data must be retained for a minimum of five years.",
    "If the root cause cannot be identified, the deviation management process shall proceed to a formal CAPA.",
    "The analyst shall review the results, calculations, and acceptance criteria to determine whether the result is reportable in accordance with applicable procedures.",
    "As described in Section 5.3, the final investigation report shall be approved by the quality unit."
]

len(sentences)
```

```python
# Sentence length (word count)

def word_count(sentence):
    return len(sentence.split())

for i, s in enumerate(sentences, start=1):
    print(f"S{i}: {word_count(s)} words")
```

```python
def is_long(sentence):
    return word_count(sentence) > 30

for i, s in enumerate(sentences, start=1):
    print(f"S{i}: {is_long(s)}")
```

```python
import random

random.seed(42)

actions = [
    "a retest shall be performed",
    "the investigation shall be initiated",
    "the analyst shall review historical data",
    "all raw data must be retained for a minimum of five years",
    "the deviation management process shall proceed to a formal CAPA",
    "the final investigation report shall be approved by the quality unit",
]

conditions = [
    "If the test result exceeds the acceptance criteria",
    "When additional testing is required",
    "Unless otherwise specified",
    "Where applicable",
    "If the root cause cannot be identified",
    "If an equipment malfunction is suspected",
]

exceptions = [
    "However",
    "This step is not applicable to stability samples",
    "Otherwise",
]

cross_refs = [
    "in accordance with SOP-QC-012",
    "as described in Section 5.3",
    "in accordance with applicable procedures",
    "per SOP-QA-101",
]

discretion = [
    "may be extended as appropriate",
    "as deemed necessary by the analyst",
    "based on professional judgment",
]

def make_sentence():
    parts = []
    # 70% chance to include a condition
    if random.random() < 0.7:
        parts.append(random.choice(conditions) + ",")
    # 30% chance to include an exception marker at the start
    if random.random() < 0.3:
        parts.insert(0, random.choice(exceptions) + ",")
    # core action
    core = random.choice(actions)
    parts.append(core)
    # 50% chance to include cross-reference
    if random.random() < 0.5:
        parts.append(random.choice(cross_refs))
    # 40% chance to include discretionary language
    if random.random() < 0.4:
        parts.append(random.choice(discretion))
    sentence = " ".join(parts).strip()
    # add period if missing
    if not sentence.endswith("."):
        sentence += "."
    return sentence

synthetic_sentences = [make_sentence() for _ in range(80)]

synthetic_sentences[:5], len(synthetic_sentences)
```

```python
import pandas as pd
import re

def has_obligation(sentence):
    return bool(re.search(r"\b(shall|must|should)\b", sentence.lower()))

def has_conditional(sentence):
    return bool(re.search(r"\b(if|when|unless|where applicable)\b", sentence.lower()))

def has_exception(sentence):
    return bool(re.search(r"\b(however|not applicable|otherwise)\b", sentence.lower()))

def has_cross_ref(sentence):
    return bool(re.search(r"\b(section|sop|in accordance with|per sop)\b", sentence.lower()))

def is_long(sentence):
    return len(sentence.split()) > 30

df = pd.DataFrame({
    "sentence": synthetic_sentences,
    "obligation": [has_obligation(s) for s in synthetic_sentences],
    "conditional": [has_conditional(s) for s in synthetic_sentences],
    "exception": [has_exception(s) for s in synthetic_sentences],
    "cross_reference": [has_cross_ref(s) for s in synthetic_sentences],
    "long_sentence": [is_long(s) for s in synthetic_sentences],
})

df.head()
```
